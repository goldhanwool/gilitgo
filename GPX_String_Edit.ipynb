```python
import glob
import os.path

files = glob.glob('*.gpx') # 특정확장자를 가진 목록뽑기
for x in files:
    if not os.path.isdir(x): # 파일만 걸러내기
        filename = os.path.splitext(x) # 확장자 ;[./'./ㅣ, /;/.,l m[';=
        # print (filename[0]) # 확인 주석
        os.rename(x, filename[0]  + '.txt') # 순수파일명에 txt확장자 붙이기
```

```python
file_name = "sample.txt"
final_file = "sample_edit.txt"

f = open(file_name, encoding="UTF8")
string = f.readlines()
a = list(string)
#print(a)
data = []
for i in a:
#    i.split(">")
    data.append(i)    
# data
# ",".join(data)

# Trk태그만 뽑기
trk = []
for i in data:
    if "lat" in i:
        trk.append(i)
        
        
# 태그 내에서 lat 좌표 분리를 위한 문자열 분리
lat = []
for i in trk:
    data = i.split(' ')
    for i in data:
        print(i)
        if "lat" in i:
            lat.append(i)
            #print(lat)


#lon 좌표 분리를 위한 문자열 분리
lon = []
for i in trk:
    data = i.split(' ')
    for i in data:
        print(i)
        if "lon" in i:
            lon.append(i)
            print(lon)
```

```python
# 데이터 프레임으로 재구조화
import pandas as pd
import numpy as np

gpx = [ x for x in zip(lat,lon)] 
df = pd.DataFrame(gpx, columns=['k', 'v'])
df


# 새로운 컬럼을 통해서 좌표만 추출
df['grp'] = df.v.str.split('>').str[0]
df['lon'] = df.grp.str.split('<').str[0]
df = df.drop(['v', 'grp'], axis = 1)
df
```

```python
import numpy as np

# 판다스 행을 리스트로 반환하기
ls = np.array(df['k'].tolist())
#list(a)
ls = df.values.tolist()
#list(a)

lat_lon=[]
for i in ls:
    print(i)
    trk_tag = str(' '.join(i))
    trk_tag = '<trkpt ' + trk_tag + '>' + '</trkpt>'
    lat_lon.append(trk_tag)
#lat_lon
```

```python
#파일을 열고 쓴다. 리스트에 담긴 내용을

with open('sample.txt', 'w') as f:    
    for line in lat_lon:
        f.write(line)

# 맨 첫부분 뒷부분 완성하고 파일로 쓰기
f = open('sample.txt', mode='rt', encoding='utf-8')
for line in f:
    gpx = '<gpx><trk><trkseg>' + line + '</trkseg></trk></gpx>'
    #p.append(g)
    #print(gpx)

f = open(final_file, mode='wt', encoding='utf-8') 
f.write(gpx)
```
