## 위경도 웨이포인트 거리 구하기

```python
import glob
import os.path

# 파일 불러오기
files = glob.glob('./files/*.txt') # 특정확장자를 가진 목록뽑기
file_name = []
for x in files:
    if not os.path.isdir(x): # 파일만 걸러내기
        filename = os.path.splitext(x) # 확장자 ;[./'./ㅣ, /;/.,l m[';=
        file_name.append(filename)
        print(file_name)
        # [];'l., , 와 순수파일명 구분하기
        # print (filename[0]) # 확인 주석
        os.rename(x, filename[0]  + '.txt') # 순수파일명에 txt확장자 붙이기
```

```python
file = []
for i in file_name:
    file.append(i[0])        

#코스이름들 전처리 - files\\ 글자 제거
file_list =[]
for i in file:
    a = i.replace("./files\\", "")
    file_list.append(a)  

file_list
```

```python
import glob
import os.path
from scipy.spatial import distance
import pandas as pd
import numpy as np

all_info1 = pd.DataFrame()
file_list_1 = []

# 코스제목들 리스트 담아 만들기
for name in file_list: 
    print(name)
    f = open("./files/%s.txt" %name, encoding='utf-8')
    string = f.readlines()
    a = list(string)
#     print(a)

# 파일이름과 확장자 분리하기
    for i in a:
        data = i.split(' ')
#         print(data)

# 위도값추출
    key = []
    for i in data:
        if i.startswith("lat") :
            key.append(i)
    
# 경도값추출
    value = []
    for j in data:
        if j.startswith("lon") :
            value.append(j)
 
# 데이터프레임으로 만들기
    gpx = [ x for x in zip(key,value)] 
    df = pd.DataFrame(gpx, columns=['lat', 'v'])

# 데이터프레임 컬럼에 문자열 분리하기
    df['grp'] = df.v.str.split('>').str[0]
    df['lon'] = df.grp.str.split('<').str[0]
    df = df.drop(['v', 'grp'], axis = 1)

    
    gps_split = df["lat"].str.split("=")
    df["lat"] = gps_split.str.get(1)

    
    gps_split2 = df["lon"].str.split("=")
    df["lon"] = gps_split2.str.get(1)

# 추출된 위경도 값을 담은 리스트 만들기    
    gps = []
    for x, y in zip(df['lat'], df['lon']) :
        x = x.replace('"','')
        y = y.replace('"','')
        z = [float(x),float(y)]
        gps.append(z)

# 위경도 포인트별 계산하기      
    result =[]
    for i in range(len(gps)):
        x1 = gps[i-1][0]
        y1 = gps[i-1][1]    
        x2 = gps[i][0]
        y2 = gps[i][1]
        result.append(round(distance.euclidean((x1, y1), (x2, y2)),5)*100)
#         print(len(result))

# 50미터 미만이면 리스트에 담기
    a = []
    
    for i in result:
        if i >= 0.05 :
            a.append(i)
    c = len(a)
    print(c)

# 코스이름과 50미터 이상되는 웨이포인트 개수 담기   
    file_list_2 =[]
    all_info ={}
    if c == 0:
        file_list_1.append(name)
    else:
        file_list_2.append(name)
        all_info = {name:c}

#딕셔너리에 담아 데이터 프레임으로 만들기        
    all_info0 = pd.DataFrame(list(all_info.items()), columns=['name', 'count'])       
    all_info1 =  pd.concat([all_info1, all_info0])
all_info1
```
